## 2. Introduction

A classical computer stores information using bits, which can only have one of two values:

. 0

. 1

a quantum computer stores information using qubits. Unlike classical bits, qubits can exist in a superposition, meaning they can represent both 0 and 1 simultaneously until they are measured.

## 3. Classical Bit Vs Qubit

| Classical Bit | Quantum Bit (Qubit) |
|----------------|--------------------|
| 0 or 1 | α|0⟩ + β|1⟩ |
| Deterministic | Probabilistic |
| No Superposition | Supports Superposition |
| No Entanglement | Can be Entangled |

## 8. Applying Basic Quantum Gates

## X Gate (Bit Flip)

The X gate flips the state of a qubit.

|0> -> |1>

|1> -> |0>

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.x(0)

print(qc.draw())
```

Explanation:

The X gate behaves like a classical NOT gate by flipping the qubit from |0> to |1>.

## 9. Hadamard (H) Gate

The Hadamard gate creates a superposition.



```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.h(0)

print(qc.draw())
```

Explanation:

The H gate transforms the qubit into superposition state, allowing it to have probabilities of being measured as both |0> and |1>.

## 10. Z Gate (phase flip)

|0> -> |0>

|1> -> |-1>



```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.z(0)

print(qc.draw())
```

Explanation:

The Z gate changes the phase of the |1> component without changing the measurement probabilities.

## 11. Y Gate

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.y(0)

print(qc.draw())
```

Explanation:

The Y gate combines a bit flip and a phase flip

## 12. Applying Multiple Gates

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.h(0)
qc.z(0)
qc.x(0)

print(qc.draw())
```
