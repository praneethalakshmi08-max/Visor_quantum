## 2. Introduction

Quantum states are represented mathematically using vectors. While these vectors are precise, they can be difficult for beginners to interpret.

The Bloch Sphere provides a three-dimensional graphical representation of a single qubit, making it easier to visualize quantum states and understand how quantum gates change those states.

It is one of the most widely used educational tools in quantum computing.

## 3. What is the Bloch Sphere?

The Bloch Sphere is a unit sphere where every point on its surface represents a valid quantum state of a single qubit.

Special points include:

* North Pole -> |0>

* South Pole -> |1>

* Points on the equator -> superposition states

This visualization helps learners understand concepts that cannot be seen directly.

## 4. Bloch Sphere Axes

The Bloch Sphere has three axes:

* X-axis: Represents rotations caused by the X gate.

* Y-axis: Represents rotations caused by the Y gate.
* Z-axis: Represents rotations caused by the Z gate.

These axes help visualize how different quantum gates affect a qubit.

## 6. Effect of the Hadamard Gate

The Hadamard gate creates a superposition.




````python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.h(0)

print(qc.draw())
````

## Output

```text

   ┌───┐
q: ┤ H ├
   └───┘

```

## Explanation

The Hadamard gate moves the qubit from the North Pole to the equator of the Bloch Sphere, representing an equal probability of measuring |0> or |1>.

## 10. Visualizing Rotations

Quantum gates perform rotations on the Bloch Sphere.

| Gate | Rotation                           |
| ---- | ---------------------------------- |
| X    | 180° about X-axis                  |
| Y    | 180° about Y-axis                  |
| Z    | 180° about Z-axis                  |
| H    | Moves the qubit into superposition |

These rotations provide an intuitive way to understand quantum operations.
